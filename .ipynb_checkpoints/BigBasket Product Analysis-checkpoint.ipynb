```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('BigBasket Products.csv')
```

```python
df.info()
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['index'])
```

```python
df_cleaned = df.drop_duplicates()
```

```python
df_cleaned = df_cleaned.dropna(subset = ['product', 'brand'])
```

```python
df_cleaned['rating'] = df_cleaned['rating'].fillna(df_cleaned['rating'].mean())
```

```python
df_cleaned['description'] = df_cleaned['description'].fillna('No description available')
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.head(5)
```

```python
df_cleaned['discount'] = df_cleaned['market_price'] - df_cleaned['sale_price']
```

```python
df_cleaned['discount_percent'] = df_cleaned['discount'] / df_cleaned['market_price'] * 100
```

```python
df_cleaned.head(5)
```

```python
df_cleaned.columns.str.lower().str.strip()
```

```python
# Count products per category
category_counts = df_cleaned['category'].value_counts().head(10)

plt.figure(figsize = (10,5))
sns.barplot(x = category_counts.values, y = category_counts.index, palette = 'Set2')
plt.title("Top 10 Product Categories by Count")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
# Focus on top 3 categories
top_categories = category_counts.index[:3]

# Group by sub-category
subcat_counts = df_cleaned[df_cleaned['category'].isin(top_categories)] \
    .groupby(['category', 'sub_category'])['product'].count().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=subcat_counts, x='product', y='sub_category', hue='category', dodge=False)
plt.title("Sub-Category Distribution in Top 3 Categories")
plt.xlabel("Number of Products")
plt.ylabel("Sub-Category")
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
#Distribution of sales price

plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned['sale_price'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Number of Products")
plt.xlim(0, df_cleaned['sale_price'].quantile(0.95))  # limit to 95th percentile to avoid extreme outliers
plt.tight_layout()
plt.show()
```

```python
#Average sales price per category
avg_price_cat = df_cleaned.groupby('category')['sale_price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_price_cat.values, y=avg_price_cat.index, palette="coolwarm")
plt.title("Average Sale Price by Category")
plt.xlabel("Average Sale Price")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
#Top 10 products with highest discount
top_discounts = df_cleaned.sort_values(by='discount_percent', ascending=False)[['product', 'brand', 'sale_price', 'market_price', 'discount_percent']].head(10)
print(top_discounts)
```

```python
#Average discount percentage by categories
avg_discount_cat = df_cleaned.groupby('category')['discount_percent'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_discount_cat.values, y=avg_discount_cat.index, palette="viridis")
plt.title("Average Discount % by Category")
plt.xlabel("Average Discount %")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
#Distribution of product rating
plt.figure(figsize=(8, 4))
sns.histplot(df_cleaned['rating'], bins=30, kde=True, color='orange')
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.tight_layout()
plt.show()
```

```python
# Top 5 rated products
top_rated = df_cleaned.sort_values(by='rating', ascending=False)[['product', 'brand', 'rating']].head(5)
print("Top Rated Products:")
print(top_rated)
```

```python
# Lowest 5 rated products
low_rated = df_cleaned[df_cleaned['rating'] > 0].sort_values(by='rating')[['product', 'brand', 'rating']].head(5)
print("\nLowest Rated Products:")
print(low_rated)
```

```python
#Average rating per category
avg_rating_cat = df_cleaned.groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=avg_rating_cat.values, y=avg_rating_cat.index, palette="Spectral")
plt.title("Average Rating by Category")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
#Top 10 brands by product count
top_brands = df_cleaned['brand'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="Set3")
plt.title("Top 10 Brands by Number of Products")
plt.xlabel("Number of Products")
plt.ylabel("Brand")
plt.tight_layout()
plt.show()
```

```python
#Top 10 brands with average rating
avg_rating_brand = df_cleaned.groupby('brand')['rating'].agg(['mean', 'count'])
top_avg_rating = avg_rating_brand[avg_rating_brand['count'] >= 20].sort_values(by='mean', ascending=False).head(10)

top_avg_rating = top_avg_rating.rename(columns={'mean': 'average_rating', 'count': 'product_count'})

print("Top 10 Brands by Average Rating (min 20 products):")
print(top_avg_rating)
```

```python
#Top 10 brands by average discount percent
avg_discount_brand = df_cleaned.groupby('brand')['discount_percent'].agg(['mean', 'count'])
top_avg_discount = avg_discount_brand[avg_discount_brand['count'] >= 20].sort_values(by='mean', ascending=False).head(10)

top_avg_discount = top_avg_discount.rename(columns={'mean': 'average_discount', 'count': 'product_count'})

print("Top 10 Brands by Average Discount % (min 20 products):")
print(top_avg_discount)
```

```python
df_cleaned.to_csv('BigBasket.csv', index = False)
```
